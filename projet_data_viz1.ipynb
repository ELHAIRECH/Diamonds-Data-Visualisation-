# ***Visualisation de la Data***



1.   **Price**





```

sas=sns.distplot(bij_df['price'],bins=20,color='r')
sas.set_title('Price Distribution')
sas.set_facecolor('y')
```

```
desc_price=bij_df['price'].describe()
desc_price
```

```
bij_df['price'].var()
```



2.   ***Color***



```
v=bij_df['color'].value_counts().plot(kind='bar')
v.set_title('Color distribution')
v.set_xlabel('Colors')
v.set_ylabel('Count')
v.set_facecolor('y')
```

```
desc_Color=bij_df['color'].describe()
desc_Color
```



3.   **Cut**

```
bij_df=bij_df.sort_values(by='carat')
bij_arr=bij_df.values
```

```
x1=bij_arr[:,1]
y1=bij_arr[:,7]
fig=plt.figure(figsize=(5,4))
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('carat')
ax.set_ylabel('Price')
ax.plot(x1,y1,label='Price en fct de carat',color='r')
ax.legend(loc=0)
ax.set_facecolor('y')

```

```
fig1=plt.figure(figsize=(5,4))
ax=fig1.add_axes([0,0,1,1])
ax.set_xlabel('carat')
ax.set_ylabel('Price')
ax.scatter(x1,y1,label='Price en fct de carat',c='r',edgecolors='b')
ax.legend(loc=0)
```
